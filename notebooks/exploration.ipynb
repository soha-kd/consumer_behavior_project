This notebook contains the exploratory data analysis for the consumer behavior dataset.  
The goal is to better understand the data before building a multiclass classification model.

The target variable is `shopping_preference`, with the classes:
- Store
- Online
- Hybrid

## Import libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Load the dataset

```python
df = pd.read_csv("../data/online vs store shopping dataset.csv")
df.head()
```

## 2. Check dataset shape

```python
print("Rows and columns:", df.shape)
```

```python
df.info()
```

## 4. Summary statistics

```python
df.describe()
```

```python
df.describe(include="object")
```

## 5. Missing values check

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
for col in df.columns:
    print(col, "->", df[col].nunique())
```

## 7. Target variable distribution

```python
df["shopping_preference"].value_counts()
##imbalanced
```

```python
df["shopping_preference"].value_counts(normalize=True)
```

```python
df["shopping_preference"].value_counts().plot(kind="bar")
plt.title("Shopping Preference Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

## 9. Numerical feature distributions

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns

df[num_cols].hist(figsize=(16,14), bins=20)
plt.suptitle("Numerical Feature Distributions")
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.select_dtypes(include=["int64", "float64"]).corr(), cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Age vs shopping preference

```python
sns.boxplot(x="shopping_preference", y="age", data=df)
plt.title("Age vs Shopping Preference")
plt.show()
```

```python
sns.boxplot(x="shopping_preference", y="monthly_income", data=df)
plt.title("Income vs Shopping Preference")
plt.show()
```
